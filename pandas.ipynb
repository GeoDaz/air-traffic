```python
print(
    "L'aéroport de départ le plus emprunté est :", 
    airports[
        airports['faa'] == flights['origin'].value_counts().idxmax()
    ]['name'].iloc[0]
)
```

```python
print("Les 10 destinations les plus prisées : ")
most_ten_dest = flights['dest'].value_counts().head(10)

name_ten_dest = []
for faa in most_ten_dest.keys():
    name_ten_dest.append(airports[airports['faa'] == faa]['name'].iloc[0])

pandas.concat([pandas.Series(name_ten_dest), pandas.Series(most_ten_dest.index), pandas.Series(most_ten_dest.values)], axis=1, keys=['Destination', 'FAA', 'Nbr de vols'])
```

```python
print("Les 10 destinations les moins prisées : ")
less_ten_dest = flights['dest'].value_counts(ascending=True).head(10)

name_ten_dest = []
for faa in less_ten_dest.keys():
    name_ten_dest.append(airports[airports['faa'] == faa]['name'].iloc[0])

pandas.concat([pandas.Series(name_ten_dest), pandas.Series(less_ten_dest.index), pandas.Series(less_ten_dest.values)], axis=1, keys=['Destination', 'FAA', 'Nbr de vols'])
```

```python
print("Les 10 avions qui ont le plus décollé : ")
# first row is a sum
flights['tailnum'].value_counts()[1:11].to_frame(name='Nbr de vols')
```

```python
# on observe que tous les avions liés à des vols ne sont pas déclarés dans planes.csv
planes_tailnum = planes.tailnum.values
most_took_off = flights['tailnum'].value_counts()[1:11].index
for x in most_took_off:
    for y in planes_tailnum:
      if x == y:
        print(planes[planes['tailnum'] == x])
```

QUESTION 5 - Trouver combien chaque compagnie a desservi de destination ; combien chaque compagnie a desservie de destination par aéroport d’origine. Réaliser les graphiques adéquats qui synthétisent ces informations ?

```python
print("Nombre de destinations desservies par compagnie : ")
# print('dest_per_airline', flights.groupby('carrier').count().dest)

dest_per_airline = pandas.Series(flights.groupby('carrier').count().dest.values)
pandas.concat([airlines, dest_per_airline], axis=1, keys=['Compagnies', 'Destinations'])
```

```python
print(pandas.Series(flights.groupby('carrier').count().dest).plot.bar(title='Nombre de destinations desservies par compagnie', ylabel='Destinations', xlabel='Compagnies'))
```

```python
print("Nombre de destinations desservies par aeroport d'origine pour chaque compagnie : ")
# print('dest_per_origin', flights.groupby(['carrier', 'origin']).count().dest)

pandas.Series(flights.groupby(['carrier', 'origin']).count().dest).to_frame('Destinations')
```

```python
print(pandas.Series(flights.groupby(['carrier', 'origin']).count().dest).plot.bar(title='Destinations desservies par aeroport d\'origine pour chaque compagnie', ylabel='Nombre de destinations', xlabel='Destination par compagnie'))
```

```python
# zoom sur une compagnie : 9E
print(pandas.Series(flights.groupby(['carrier', 'origin']).count().dest.loc['9E']).plot.bar(title='Destinations desservies par la compagnie 9E', ylabel='Nombre de destinations', xlabel='Aeroport d\'origine'))
```
